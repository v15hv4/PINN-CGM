```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adult001 = pd.read_csv("results/2023-04-12_03-10-46/adult#001.csv")
```

```python
adult001
```

```python
t = adult001["Time"].to_numpy()
CGM = adult001["CGM"].to_numpy()
CHO = adult001["CHO"].to_numpy()
insulin = adult001["insulin"].to_numpy()

plt.figure(figsize=(10, 10))

plt.subplot(312)
plt.plot(t, CHO, "y")
plt.xticks([], [])

plt.subplot(313)
plt.plot(t, insulin, "g")
plt.xticks([], [])

plt.subplot(311)
plt.plot(t, CGM, "r")
plt.xticks([], [])

plt.tight_layout()
plt.show()
```
