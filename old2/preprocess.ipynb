```python
import numpy as np
import seaborn as sns

from utils import visualize
from data import C3RData, interpolate
```

```python
# better plots
sns.set()
```

```python
# load dataset
c3r = C3RData(data_dir="data")
```

```python
# get data for a certain case
deident_id = 2
from_dt = f"2014-05-05 00:00:00"
to_dt = f"2014-05-08 00:00:00"

cgm_data, meal_data, insulin_data = c3r.get_case(deident_id, from_dt, to_dt)
t, cgm, meal, insulin = interpolate(cgm_data, meal_data, insulin_data)
```

```python
# visualize raw data
visualize(
    (cgm_data.LocalDtTm, cgm_data.CGM), 
    (meal_data.LocalDtTm, meal_data.MealSize), 
    (insulin_data.LocalDeliveredDtTm, insulin_data.DeliveredValue), 
    fmt=".",
)
```

```python
# visualize interpolated data
visualize(
    (t, cgm), 
    (t, meal), 
    (t, insulin), 
    fmt="-",
)
```
